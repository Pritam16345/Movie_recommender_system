```python
# I have to make this full format in this form :-
# ['Action','Adventure','Fantasy','Science Fiction']
```

```python
import ast

def convert(obj) :
    L = []
    for i in (ast.literal_eval(obj)):
        L.append(i['name'])
    return L

# the problem here arises that when genres list is extracted in the obj, it is coming in the form of strings(not in dictionary form).
# So we convert that into list by using literal_eval
```

```python
movies['genres'].apply(convert)
```

```python
movies['genres'] = movies['genres'].apply(convert)
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies['cast'][0]
```

```python
import ast

def convert3(obj) :
    L = []
    count = 0
    for i in (ast.literal_eval(obj)):
        if (count != 3) :
            L.append(i['name'])
            count+=1
        else :
            break
    return L
```

```python
movies['cast'] = movies['cast'].apply(convert3)
movies.head()
```

```python
movies['crew'][0]
```

```python
import ast

def fetch_director(obj):
    L = []
    if isinstance(obj, str):  # if it's a string, convert it
        try:
            obj = ast.literal_eval(obj)
        except:
            return L  # if conversion fails, return empty list
            
    for i in obj:
        if i.get('job') == 'Director':
            L.append(i.get('name'))
    return L
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.head()
```

```python
movies['overview'][0]
```

```python
movies['overview'] = movies['overview'].apply(lambda x : x.split())
```

```python
movies.head()
```

```python
# Now another problem arises that we need to remove spaces in between names or anywhere like :-
# For eg - Science Fiction to ScienceFiction
```

```python
movies['genres'] = movies['genres'].apply(lambda x : [i.replace(" ","") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x : [i.replace(" ","") for i in x])
movies['cast'] = movies['cast'].apply(lambda x : [i.replace(" ","") for i in x])
movies['crew'] = movies['crew'].apply(lambda x : [i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies.head()
```

```python
new_df = movies[['movie_id','title','tags']]
```

```python
new_df.head()
```

```python
#converting all the list into string
new_df['tags'] = new_df['tags'].apply(lambda x : " ".join(x))
```

```python
new_df.head()
```

```python
new_df['tags'][0]
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x : x.lower())
```

```python
new_df.head()
```

```python
#---------------------------------------------------------------------------------------------
```

```python
# Vectorization
# We have to find the similarties between the tags of different movies. So we have to convert the tags of movies into vectors
# Then every movies will become a vector. Then while recommending, the user will give a movie, then we will find the closest vectors to the vector of that particular movie.
# So we have to convert text to vector which is known as text vectorization. Here we will use bag of words technique(simple).
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000, stop_words='english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors
```

```python
vectors.shape
```

```python
vectors[0]
```

```python
cv.get_feature_names_out()
```

```python
# Now here words like'activity' 'activities', 'movie''movies' should be same.
# So we apply stemming - Stemming is a text normalization technique used in Natural Language Processing (NLP) to reduce words to their root or base form.
# The idea is to strip suffixes (like "ing", "ed", "s") so that similar words are treated the same.
```

```python
from nltk.stem import PorterStemmer
ps = PorterStemmer()
# example here
ps.stem('loved')
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
cv.get_feature_names_out()
```

```python
# We will not find Euclidean distance(distance between tips). We will use cosine similarity(theta between two vectors)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity[0]
# This gives the distance betweent the first movie with all other ones.
```

```python
similarity.shape
```

```python
# to find the index number of a movie
new_df[new_df['title'] == 'Batman Begins'].index[0]
```

```python
# Now the problem is if we sort the movies in descending order, the indices will mixup. So we have to do something to hold the indexes.
#For that reason we use the enumerate function
list(enumerate(similarity[0]))
# Here it will print with indexes.
```

```python
sorted(list(enumerate(similarity[0])), reverse=True,key=lambda x:x[1])
# key=lambda x:x[1] We used this so that we can sort with respect to distance and not indexes
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    
    for i in movies_list:
        print(i)
```

```python
recommend('Avatar')
```
